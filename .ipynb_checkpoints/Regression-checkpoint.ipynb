```python
def encode(df_input, encode_column, nulls='ignore'):
    enc = OneHotEncoder(handle_unknown=nulls)
    return pd.DataFrame(enc.fit_transform(df_input[[encode_column]]).toarray(), columns = enc.get_feature_names_out([encode_column]))
        
df_y = df[['OPP_ID','ACCOUNT_ID', 'REP_ID']]
df_y['WON'] = (df.OUTCOME == 'Closed Won').astype(int)

df_x = encode(df, 'OFFICE')
df_x = df_x.join(encode(df, 'PRIMARY_TYPE'))
df_x = df_x.join(encode(df, 'SECONDARY_TYPE'))

stage_ind = ['DATE_OF_QUALIFYING', 'DATE_OF_EVALUATION', 'DATE_OF_PROCUREMENT_NEGOTIATIONS', 'DATE_OF_VERBAL_PENDING']
df_x[stage_ind] = df[stage_ind].notnull().astype(int)
```

```python
date_columns = ['CREATED_DATE', 'CLOSE_DATE', 'DATE_OF_QUALIFYING', 'DATE_OF_EVALUATION', 'DATE_OF_PROCUREMENT_NEGOTIATIONS', 'DATE_OF_VERBAL_PENDING']
for col in date_columns:
    df[col] = pd.to_datetime(df[col])
```

```python
scaler = StandardScaler()
def sft(input, ss=scaler):
    return ss.fit_transform(np.array(input).reshape(-1,1))

next_event = ['CLOSE_DATE','DATE_OF_QUALIFYING', 'DATE_OF_EVALUATION', 'DATE_OF_PROCUREMENT_NEGOTIATIONS', 'DATE_OF_VERBAL_PENDING']
last_event = ['CREATED_DATE','DATE_OF_QUALIFYING', 'DATE_OF_EVALUATION', 'DATE_OF_PROCUREMENT_NEGOTIATIONS', 'DATE_OF_VERBAL_PENDING']

df_x['OPPORTUNITY_OPEN_DAYS'] = sft((df['CLOSE_DATE'] - df['CREATED_DATE']).dt.days)
df_x['FIRST_EVENT_DAYS'] = sft((df.apply(lambda x: min([x[i] for i in next_event]), axis=1) - df['CREATED_DATE']).dt.days)
df_x['LAST_EVENT_DAYS'] = sft((df['CLOSE_DATE'] - df.apply(lambda x: max([x[i] for i in next_event]), axis=1)).dt.days)

print(df_x[['FIRST_EVENT_DAYS','LAST_EVENT_DAYS']].isnull().sum())
```

```python
X = df_x
y = df_y[['WON']].astype(int)

# Split the data into training and testing sets with stratification
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=17)

# Initialize Logistic Regression model with L1 regularization to reduce dimensional space / perform feature selection
logreg_model = LogisticRegression(penalty='l1', solver='liblinear', random_state=17)

# Fit the model on the training data
logreg_model.fit(X_train, y_train)

# Predict on the test data
y_pred = logreg_model.predict(X_test)

# Calculate accuracy on the test set
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy on the test set: {accuracy:.2f}')

# Print the coefficients
coefficients = logreg_model.coef_[0]
intercept = logreg_model.intercept_

print("Intercept:", intercept)
coef_df = pd.DataFrame({'Feature': df_x.columns, 'Coefficient': coefficients})
print("Coefficients with Odds Ratios:")
print(coef_df[abs(coef_df.Coefficient) > 0].reset_index()[0:20])
print(coef_df[abs(coef_df.Coefficient) > 0].reset_index()[20:40])
print(coef_df[abs(coef_df.Coefficient) > 0].reset_index()[40:60])
print(coef_df[abs(coef_df.Coefficient) > 0].reset_index()[60:80])
```

```python
# Output confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)

# Output precision and recall
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)

print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"Accuracy: {accuracy:.2f}")
```
